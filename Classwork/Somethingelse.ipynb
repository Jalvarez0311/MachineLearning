```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

iris = load_iris()
X = iris.data[:, 2:]
y = iris.target
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=46)
tree_clf.fit(X,y)

tree_clf.predict([[5, 1.5]])
```

```python
tree_clf.predict_proba([[5,1.5]])
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

tree.plot_tree(tree_clf, class_names=iris.target_names, feature_names=iris.feature_names[2:], rounded=True, filled=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np

np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m, 1) / 10

tree_reg = DecisionTreeRegressor(max_depth=2, random_state=42)
tree_reg.fit(X,y)

tree.plot_tree(tree_reg, rounded=True, filled=True)
plt.show()
```
